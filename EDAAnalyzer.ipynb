```python
import numpy as np
import pandas as pd

# Fonksiyon: HHMM formatındaki değeri dakika cinsine çevirir (NaN ise NaN döndürür)
def hhmm_to_minutes(time_val):
    if pd.isna(time_val):
        return np.nan
    time_str = str(int(time_val)).zfill(4)  # 4 haneli string
    hour = int(time_str[:2])
    minute = int(time_str[2:])
    return hour * 60 + minute

# CRSDepTime için işlemler:
df['CRSDepTime_minutes'] = df['CRSDepTime'].apply(hhmm_to_minutes)
df['CRSDepTime_str'] = df['CRSDepTime'].astype(str).str.zfill(4)  # Örneğin "1640" gibi
df['CRSDepTime_time'] = pd.to_datetime(df['CRSDepTime_str'], format='%H%M').dt.time

# DepTime için:
# NaN değerleri olduğu gibi bırakıp, diğer geçerli değerleri "HHMM" formatında 4 haneli string'e çevirecek fonksiyon:
def convert_dep_time(x):
    if pd.isna(x):
        return np.nan
    return f"{int(x):04d}"

df['DepTime_minutes'] = df['DepTime'].apply(hhmm_to_minutes)
df['DepTime_str'] = df['DepTime'].apply(convert_dep_time)
df['DepTime_time'] = pd.to_datetime(df['DepTime_str'], format='%H%M', errors='coerce').dt.time

# DepTime sütununda sıfır değeri kontrolü (eğer sıfır varsa, ilgili mesaj yazdırır)
if (df['DepTime'] == 0).any():
    print("DepTime sütununda sıfır değeri bulunuyor.")
else:
    print("DepTime sütununda sıfır değeri bulunmuyor.")
```

```python
import numpy as np
import pandas as pd

# 1) WheelsOff değerini dakika cinsine dönüştüren fonksiyon
def hhmm_to_minutes(time_val):
    if pd.isna(time_val):  # NaN ise direkt NaN dön
        return np.nan
    # Değer tam sayı olmalı, 4 haneli string'e dönüştür
    time_str = f"{int(time_val):04d}"
    hour = int(time_str[:2])
    minute = int(time_str[2:])
    return hour * 60 + minute

# 2) WheelsOff değerini "HHMM" formatında 4 haneli string'e çeviren fonksiyon
def convert_time_str(x):
    if pd.isna(x):
        return np.nan
    return f"{int(x):04d}"

# WheelsOff sütununu dakika cinsine çevir
df['WheelsOff_minutes'] = df['WheelsOff'].apply(hhmm_to_minutes)
df['WheelsOn_minutes'] = df['WheelsOn'].apply(hhmm_to_minutes)

# WheelsOff sütununu time nesnesine çevir
df['WheelsOff_str'] = df['WheelsOff'].apply(convert_time_str)
df['WheelsOff_time'] = pd.to_datetime(df['WheelsOff_str'], format='%H%M', errors='coerce').dt.time

df['WheelsOn_str'] = df['WheelsOn'].apply(convert_time_str)
df['WheelsOn_time'] = pd.to_datetime(df['WheelsOn_str'], format='%H%M', errors='coerce').dt.time
```

```python
# WheelsOff sütunundaki eksik değerleri kontrol edip, yeni bir indicator sütunu oluşturuyoruz:
df['WheelsOff_missing'] = df['WheelsOff'].isna().astype(int)
df['WheelsOn_missing'] = df['WheelsOn'].isna().astype(int)


# Sonuç olarak, eğer WheelsOff değeri NaN ise, WheelsOff_missing sütunu 1, aksi halde 0 olacaktır.
print(df[['WheelsOff', 'WheelsOff_missing']].head())
print(df[['WheelsOn', 'WheelsOn_missing']].head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Gereksiz sütunları çıkararak yeni dataframe oluştur
df_new = df.drop(columns=['Quarter', 'DepTime', 'CRSArrTime', 'DepDelay', 'ArrDelayMinutes', 'WheelsOn', 'CRSElapsedTime', 'ActualElapsedTime', 'AirTime'])

# Yeni bir feature oluştur: NetDelayImpact (Varış Gecikmesi - Kalkış Gecikmesi)
df_new['NetDelayImpact'] = df_new['ArrDelay'] - df_new['DepDelayMinutes']

# Yeni feature: EarlyArrivalImpact (Eğer uçak erken inmişse 1, yoksa 0)
df_new['EarlyArrivalImpact'] = df_new['ArrDelay'].apply(lambda x: 1 if x < 0 else 0)

# Yeni feature oluştur: ElapsedTimeDiff (Gerçek uçuş süresi - Planlanan uçuş süresi)
df_new['ElapsedTimeDiff'] = df['ActualElapsedTime'] - df['CRSElapsedTime']

# Yeni feature: Ortalama Hız (Speed) → Distance / ActualElapsedTime
df_new['Speed'] = df['Distance'] / df['ActualElapsedTime']

# Yeni feature: DelayEfficiency (Net gecikme süresinin uçuş süresine oranı)
df_new['DelayEfficiency'] = df_new['NetDelayImpact'] / (df_new['ElapsedTimeDiff'] + 1e-5)

# RelativeDelayImpact'i kaldırıyoruz çünkü DelayEfficiency ile aynı bilgiyi taşıyor!
# df_new.drop(columns=['RelativeDelayImpact'], inplace=True)  # Çıkardık!

# Güncellenmiş numeric sütun listesini oluştur
numeric_columns_updated = [col for col in numeric_columns if col not in 
                           ['Quarter', 'DepTime', 'CRSArrTime', 'DepDelay', 
                            'DepDelayMinutes', 'ArrDelayMinutes', 'WheelsOn', 'CRSElapsedTime', 'ActualElapsedTime', 'AirTime']]
# Yeni feature'ları listeye ekle
numeric_columns_updated.extend(['NetDelayImpact', 'EarlyArrivalImpact', 'ElapsedTimeDiff', 'Speed', 'DelayEfficiency'])

# Korelasyon matrisini yeniden oluştur
new_corr_matrix = df_new[numeric_columns_updated].corr()

# Korelasyon matrisini görselleştirme
plt.figure(figsize=(12,10))
sns.heatmap(new_corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Yeni Korelasyon Matrisi (Düzgün Sayısal Sütunlar)")
plt.show()
```

```python
import numpy as np
import pandas as pd

# Örnek: Log dönüşümlerini uyguluyoruz:
df_new['TaxiIn_log'] = np.log1p(df_new['TaxiIn'])
df_new['DepDelayMinutes_log'] = np.log1p(df_new['DepDelayMinutes'])
df_new['TaxiOut_log'] = np.log1p(df_new['TaxiOut'])
df_new['Distance_log'] = np.log1p(df_new['Distance'])

# Log dönüşümünden sonraki skewness değerlerini hesaplayalım:
skew_taxi_in = df_new['TaxiIn_log'].skew()
skew_depdelay = df_new['DepDelayMinutes_log'].skew()
skew_taxi_out = df_new['TaxiOut_log'].skew()
skew_distance = df_new['Distance_log'].skew()

# Sonuçları yazdıralım:
print("Log dönüşümü sonrası skewness değerleri:")
print("TaxiIn_log skewness:", skew_taxi_in)
print("DepDelayMinutes_log skewness:", skew_depdelay)
print("TaxiOut_log skewness:", skew_taxi_out)
print("Distance_log skewness:", skew_distance)
```

```python
import numpy as np

# Negatif değer içerebileceğini belirttiğin sütunlar (örnek)
neg_cols = [
    'ArrDelay',
    'ElapsedTimeDiff',
    'DepartureEfficiency', 
    'DelayEfficiency'
]

# Signed log transform ve skewness hesaplama
for col in neg_cols:
    if col in df_new.columns:  # sütun gerçekten varsa
        # Signed log transform
        df_new[col + '_log_signed'] = (
            np.sign(df_new[col]) * np.log1p(np.abs(df_new[col]))
        )
        
        # Skewness hesapla
        skew_val = df_new[col + '_log_signed'].dropna().skew()
        
        print(f"{col} sütununa signed log transform uygulandı.")
        print(f"{col}_log_signed skewness: {skew_val:.2f}\n")
    else:
        print(f"{col} sütunu veri setinde bulunamadı, atlanıyor.\n")
```

```python
from sklearn.preprocessing import PowerTransformer

# Yeo-Johnson dönüşümü için örnek:
pt = PowerTransformer(method='yeo-johnson')
df_new['DelayEfficiency_log_signed_yj'] = pt.fit_transform(df_new[['DelayEfficiency_log_signed']])

print("DelayEfficiency_log_signed Yeo-Johnson skewness:", df_new['DelayEfficiency_log_signed_yj'].skew())

```
